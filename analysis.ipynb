```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def loadContinuous(filename):
    return (np.loadtxt(filename, dtype = [("id","i4"),("time","f8"),("start","f8"), ( "end","f8"), ("diff","f8"),("ans","i4")],comments=None, delimiter=' ' , usecols=[1,3,5,7,9,11] )).view(np.recarray)
files=!ls results/?C*
name = [x.split(".")[0].split("/")[1] for x in files]
for (n,f) in zip(name, files): 
    print(n, f)
    vars()[n]=loadContinuous(f)
```

```python

def plot1(data):
    global ax
    dat=data[data.ans==1]
    plt.plot(dat.start ,dat.diff, dat.diff/dat.time,".", color="blue")
    dat=data[(data.ans==2 )| (data.ans == 0)]
    plt.plot(dat.start, dat.diff, dat.diff/dat.time,".", color="black")
    dat=data[data.ans==3]
    plt.plot(dat.start,dat.diff, dat.diff/dat.time,".", color="red")
    ax.set_xlabel("start")
    ax.set_ylabel("diff")
    ax.set_zlabel("speed")
    ax.set_xlim([0.25, 0.75])
    ax.set_ylim([-0.5, 0.5])
    ax.set_zlim([-0.1,0.1])
    
    
```
